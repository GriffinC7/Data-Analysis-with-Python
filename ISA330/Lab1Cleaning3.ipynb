```python
#load saved data
data = pd.read_csv("AboutCleanCleaned.csv")
data.head()
```

```python
data.columns = ["duration", "date", "name", "favoriteNumber", "animalQ", "animalType", "graduationDate", "classYear", "height"]
data.head()
```

```python
print(set(data.animalType))
```

```python
#one-hot-encoding
#https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.OneHotEncoder.html

dog = []
cat = []

for animals in data.animalType: #data["animalType"]
    if type(animals) == str:
        a = animals.split(',')
        #print(a)
        if "Dog" in a:
            if "Cat" in a:
                dog.append(1)
                cat.append(1)
        
            else:
                dog.append(1)
                cat.append(0)
        elif "Cat" in a:
            cat.append(1)
            dog.append(0)
        else:
            dog.append(0)
            cat.append(0)
        
    else:
        dog.append(0)
        cat.append(0)
```

```python
print(len(dog))
print(len(cat))
print(len(data.animalType))
```

```python
data["dog"] = dog
data["cat"] = cat
data = data.drop(columns = ["animalType"])
data.head()
```

```python
data.to_csv("AboutUsCatagorized.csv")
```
