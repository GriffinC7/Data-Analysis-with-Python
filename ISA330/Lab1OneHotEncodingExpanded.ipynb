```python
import pandas as pd
import numpy as np
```

```python
#load saved data
data = pd.read_csv("AboutCleanCleaned.csv")
data.head()
```

```python
data.columns = ["duration", "date", "name", "favoriteNumber", "animalQ", "animalType", "graduationDate", "classYear", "height"]
data.head()
```

```python
print(set(data.animalType))
```

```python
#Option 1: this is what I did in class

dog = []
cat = []

for animals in data.animalType: #data["animalType"]
    if type(animals) == str:
        a = animals.split(',')
        #for item in a:  ####### This created 29 item lists. Remove this line and unindent following rows by one tab
        if "Dog" in a:
            if "Cat" in a:
                dog.append(1)
                cat.append(1)

            else:
                dog.append(1)
                cat.append(0)
        elif "Cat" in a:
            cat.append(1)
            dog.append(0)
        else:
            dog.append(0)
            cat.append(0)
                
    else:
        dog.append (0)
        cat.append (0)
```

```python
print(len(dog))
print(len(cat))
print(len(data.animalType))
```

```python
#option 3: this is what I was originally going for in class!

dog = []
cat = []

for animals in data.animalType: #data["animalType"]
    if type(animals) == str:
        a = animals.split(',')
        #for item in a:  ####### This created 29 item lists. Remove this line and unindent following rows by one tab
        if "Cat" in a:
            cat.append(1)
        else:
            cat.append(0)
            
        if "Dog" in a:
            dog.append(1)
        else:
            dog.append(0)
    
    else:
        cat.append(0)
        dog.append(0)
            
print(len(dog))
print(len(cat))
print(len(data.animalType))
```

```python
#option 4: Yet another option

dog = []
cat = []

for animals in data.animalType: #data["animalType"]
    if type(animals) == str:
        if animals == 'Dog,Cat,Reptile':
            dog.append(1)
            cat.append(1)
        elif animals == 'Dog,Fish':
            dog.append(1)
            cat.append(0)
        elif animals == 'Dog,Cat':
            dog.append(1)
            cat.append(1)
        elif animals == 'Dog':
            dog.append(1)
            cat.append(0)
        elif animals == 'Cat':
            dog.append(0)
            cat.append(1)
        else:
            dog.append(0)
            cat.append(0)
    else:
        dog.append(0)
        cat.append(0)
        
print(len(dog))
print(len(cat))
print(len(data.animalType))
```
